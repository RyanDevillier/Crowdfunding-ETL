## Deliverable 1: Extract
----
### Option 2: Use regex 

1. Import the `backer_info.csv` file into a DataFrame. 
2. Extract the "backer_id", "cf_id", "name", and "email using regular expressions."
3. Create a new DataFrame with the retrieved data.
4. Export the DataFrame as `backers_data.csv`.

```python
# Get the backers_info from the crowdfunding_info sheet. 
pd.set_option('max_colwidth', 400)
backers_info = pd.read_csv("backer_info.csv")
backers_info.head()
```

```python
# Extract the alpha-numeric "backer_id" from the backer_info column using a regex expression 
# and add it as a new column called "backer_id".

# Iterating through every row of the backers_info DataFrame to capture all of the backer_ids using a regular expression, 
# and appending these backer_ids to a list that we will place into the backers_info DataFrame as a new column.
import re
backer_id_list = []

for i in range(len(backers_info)):
    backer_id = re.findall(r'[a-z]{2}\d{3}', string = backers_info["backer_info"][i])
    
    if backer_id not in backer_id_list:
        backer_id_list.append(backer_id[0])

backers_info["backer_id"] = backer_id_list
backers_info.head(10)
```

```python
# Extract the two to four-digit "cf_id" number from the backer_info column. 
# and add it as a new column called "cf_id".
cf_id_list = []

for i in range(len(backers_info)):
    cf_id = re.findall(r'(\s+\d{2,4})', string = backers_info["backer_info"][i])
    
    if cf_id not in cf_id_list:
        cf_id_list.append(cf_id[0])

backers_info["cf_id"] = cf_id_list
backers_info.head(10)
```

```python
# Extract the name from the backer_info column and add it as a new column called "name".
name_list = []
for i in range(len(backers_info)):
    name = re.findall(r'([A-Za-z]+\s+[A-Za-z]+)', string = backers_info["backer_info"][i])
    
    if name not in name_list:
        name_list.append(name[0])
        
backers_info["name"] = name_list
backers_info.head(10)
```

```python
# Extract the email from the backer_info column and add it as a new column called "email".
emails_list = []

for i in range(len(backers_info)):
    emails = re.findall(r'[a-zA-Z]+@+[a-zA-Z]+.+[a-zA-Z]', string = backers_info["backer_info"][i])
    
    if emails not in emails_list:
        emails_list.append(emails[0])

backers_info["email"] = emails_list
backers_info.head(10)
```

```python
# Create a new DataFrame with the appropriate columns.
new_columns_backers_info = ["backer_id", "cf_id", "name", "email"]
backers_info = backers_info[new_columns_backers_info]
backers_info.head(10)
```

```python
# Export the DataFrame as a CSV file using encoding='utf8'.
backers_info.to_csv("backer_info.csv", index = False, encoding = 'utf8')
```

## Deliverable 2: Transform and Clean Data
----
1. Check the data types of the columns and convert the "cf_id" column to an integer, if necessary.
2. Split the name in the "name" column into first and last names, and add them to "first_name" and "last_name" columns in the DataFrame. 
3. Drop the "name" column in the DataFrame.
4. Place the columns in the following order; "backer_id", "cf_id", "first_name", "last_name" and "email".

```python
# Check data types.
backers_info.info()
```

```python
# Convert cf_id to an integer if necessary.
backers_info["cf_id"] = backers_info["cf_id"].astype("int64")
```

```python
# Confirming that the cf_id column now contains only values of type int64
backers_info.dtypes
```

```python
# Split the "name" column into "first_name" and "last_name" columns.
backers_info[["first_name", "last_name"]] = backers_info["name"].str.split(" ", n=1, expand = True)
backers_info.head()
```

```python
#  Drop the name column
backers_info_cleaned = backers_info.drop(["name"], axis = 1)

# Reorder the columns
new_columns_backers_info_cleaned = ["backer_id", "cf_id", "first_name", "last_name", "email"]
backers_info_cleaned = backers_info_cleaned[new_columns_backers_info_cleaned]

# Displaying the first 10 rows of the newly cleaned DataFrame:
backers_info_cleaned.head(10)
```
